```python
import pandas as pd
```

```python
df=pd.read_csv("Indian_Kids_Screen_Time.csv")
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df = df.drop("Health_Impacts",axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
plt.plot(df['Age'],df['Avg_Daily_Screen_Time_hr'])
plt.show()
```

```python
plt.pie(df["Avg_Daily_Screen_Time_hr"])
```

```python
plt.plot(df['Age'],df['Avg_Daily_Screen_Time_hr'],color='red',linestyle='--',marker='o')
plt.title("basic plot")
plt.xlabel("Age")
plt.ylabel("screen time")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.bar(df['Age'],df['Avg_Daily_Screen_Time_hr'])
plt.show()
```

```python
plt.barh(df['Age'],df['Avg_Daily_Screen_Time_hr'])
plt.show()
```
